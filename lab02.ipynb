```python
#task 1
import random
#(width,height)
blockSizes=[(5,2),(4,7),(8,3),(3,7),(2,9),(2,3)]
connections=[(0,1),(2,1),(1,3),(0,5),(3,4),(4,5)]
gridSize=25
compSize=6 #components in chromosomes
maxGen=15 #15 iteration
mutRate=0.05 #0.05-0.1
alpha=1000 #overlapping
beta=2 #wiring length error
gamma=1 #bounding area error
def randomPos(width,height):
    x=int(random.random()*(gridSize-width))
    y=int(random.random()*(gridSize-height))
    return (x,y)
def randomChromosome():
    lay=[]
    for b in blockSizes:
        lay.append(randomPos(b[0],b[1]))
    return lay
def getCenter(pos,size):
    x,y=pos
    blockWidth,blockHeight=size
    return (x+blockWidth/2,y+blockHeight/2)
def euclidean(a,b):
    dx=a[0]-b[0]
    dy=a[1]-b[1]
    ec=(dx*dx+dy*dy)**0.5
    return ec
#8.Required Output
def overlapCount(chrom):
    count=0
    for i in range(6):
        for j in range(i+1,6):
            ax,ay=chrom[i]
            aWidth,aHeight=blockSizes[i]
            bx,by=chrom[j]
            blockWidth,blockHeight=blockSizes[j]
            if not (ax+aWidth<= bx or ax >= bx+blockWidth or ay>= by+blockHeight or ay+aHeight<= by):
                count+= 1
    return count
#8.Required Output
def totalWiringLength(chrom):
    t=0.0
    for a,b in connections:
        ca=getCenter(chrom[a],blockSizes[a])
        cb=getCenter(chrom[b],blockSizes[b])
        t+= euclidean(ca,cb)
    return t
#8.Required Output
def boundingArea(chrom):
    xs,ys=[],[]
    for i in range(6):
        x,y=chrom[i]
        w,h=blockSizes[i]
        xs.append(x)
        xs.append(x+w)
        ys.append(y)
        ys.append(y+h)
    xMin=xs[0]
    xMax=xs[0]
    yMin=ys[0]
    yMax=ys[0]
    for v in xs:
        if v<xMin:
            xMin=v
        if v>xMax:
            xMax=v
    for v in ys:
        if v<yMin:
            yMin=v
        if v>yMax:
            yMax=v
    area=(xMax-xMin)*(yMax-yMin)
    return area
#fitness check
def fitness(chrom):
    return -1*(alpha*overlapCount(chrom)+
                 beta*totalWiringLength(chrom)+
                 gamma*boundingArea(chrom))
#4.Single Point Crossover
def singlePointCrossover(p1,p2):
    pt=int(random.random()*5)+1
    c1=p1[:pt]+p2[pt:]
    c2=p2[:pt]+p1[pt:]
    return c1,c2
#5.Mutation
def mutate(chrom):
    if random.random()<mutRate:
        idx=int(random.random()*6)
        w,h=blockSizes[idx]
        chrom[idx]=randomPos(w,h)
#7.GA Loop
population=[randomChromosome() for _ in range(compSize)]
for gen in range(maxGen):
    best=0
    for current in range(1,compSize):
        if fitness(population[current])>fitness(population[best]):
            best=current
    nextGen=[list(population[best])]
    #6.New Generation Creation
    while len(nextGen)<compSize:
        #3.Parent Selection
        p1=int(random.random()*compSize)
        p2=int(random.random()*compSize)
        while p1==p2:
            p2=int(random.random()*compSize)
        ch1,ch2=singlePointCrossover(population[p1],population[p2])
        mutate(ch1)
        mutate(ch2)
        nextGen.append(ch1)
        if len(nextGen)<compSize:
            nextGen.append(ch2)
    population=nextGen
best=0
for i in range(1,compSize):
    if fitness(population[i])>fitness(population[best]):
        best=i
bestChrom=population[best]
print("Best fitness:",fitness(bestChrom))
print("Total wiring length:",totalWiringLength(bestChrom))
print("Bounding box area:",boundingArea(bestChrom))
print("Overlap count:",overlapCount(bestChrom))
print("Bottom-left placements:",bestChrom)
```
